```python
for i in range(1005):#1005
    a = i 
    b = a+29
    c = str(a)+'_'+str(b)
    de_29 = pd.read_csv('/Users/riomoukiri/Desktop/CHUPROJECT/link/'+c+'.csv')
    del de_29['Unnamed: 0']
    local_max = argrelextrema(de_29["0"].values, np.greater)
    local_min = argrelextrema(de_29["0"].values, np.less)
    local_max = pd.DataFrame(local_max).T
    local_min = pd.DataFrame(local_min).T
    ex_max = []
    for i in range(len(local_max)):
        ex_max.append(local_max[0][i])
    for i in range(len(ex_max)):
        ex_max[i] = int(ex_max[i]) 
    ex_min = []
    for i in range(len(local_min)):
        ex_min.append(local_min[0][i])
    for i in range(len(ex_min)):
        ex_min[i] = int(ex_min[i])
    de_29['Minima'] = 0
    for  i in ex_min:
        a = min(de_29['linkret'][i],de_29['linkret'][i-1],de_29['linkret'][i+1])
        if a == de_29['linkret'][i-1]:
            de_29.loc[de_29['index']==i-1, 'Minima'] = 1
        elif a == de_29['linkret'][i+1]:
            de_29.loc[de_29['index']==i+1, 'Minima'] = 1
        else:
            de_29.loc[de_29['index']==i, 'Minima'] = 1
    de_29['Maxima'] = 0
    for  i in ex_max:
        a = max(de_29['linkret'][i],de_29['linkret'][i-1],de_29['linkret'][i+1])
        if a == de_29['linkret'][i-1]:
            de_29.loc[de_29['index']==i-1, 'Maxima'] = 1
        elif a == de_29['linkret'][i+1]:
            de_29.loc[de_29['index']==i+1, 'Maxima'] = 1
        else:
            de_29.loc[de_29['index']==i, 'Maxima'] = 1
    de_29.to_csv('/Users/riomoukiri/Desktop/link/'+c+'.csv')
        
        
    
        
                
    
```
